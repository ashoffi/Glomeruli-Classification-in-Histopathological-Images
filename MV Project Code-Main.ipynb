Loading Data

```python
import os
import cv2
import numpy as np
dirpath='normal'
files=os.listdir(dirpath)
norm_image_array=[]
dim=(227,227)
for file in files:
    imagepath=os.path.join(dirpath,file)
    #print(imagepath)
    image=cv2.imread(imagepath)
    image=cv2.resize(image, dim, interpolation=cv2.INTER_AREA)
    norm_image_array.append(image)
norm_image_array=np.array(norm_image_array)   
print(norm_image_array.shape)
```

```python
X_norm = np.concatenate((vf_norm_image_array, hf_norm_image_array, sr_norm_image_array, rs_norm_image_array, bc_norm_image_array))
print(X_norm.shape)
```

Model 1 : Shallow Convolutional Nueral Network

```python
# import necessary layers  
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D 
from tensorflow.keras.layers import MaxPool2D, Flatten, Dense 
from tensorflow.keras import Model
from tensorflow.keras.optimizers import RMSprop
# input

model = tf.keras.models.Sequential([
# Note the input shape is the desired size of the image 200x200 with 3 bytes color
# This is the first convolution
tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(227, 227, 3)),
tf.keras.layers.MaxPooling2D(2, 2),
# The second convolution
tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
tf.keras.layers.MaxPooling2D(2,2),
# The third convolution
tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
tf.keras.layers.MaxPooling2D(2,2),
# The fourth convolution
tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
tf.keras.layers.MaxPooling2D(2,2),
# # The fifth convolution
tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
tf.keras.layers.MaxPooling2D(2,2),
# Flatten the results to feed into a DNN
tf.keras.layers.Flatten(),
# 512 neuron hidden layer
tf.keras.layers.Dense(512, activation='relu'),
# Only 1 output neuron. It will contain a value from 0-1 where 0 for 1 class ('dandelions') and 1 for the other ('grass')
tf.keras.layers.Dense(1, activation='sigmoid')])

#model = Model (inputs=input, outputs =output)

model.compile(loss='binary_crossentropy',optimizer=RMSprop(lr=0.001),metrics='accuracy')
#model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
History = model.fit(X_train,y_train, epochs = 10, validation_data = (X_val,y_val),batch_size=128, verbose = 1)
```

```python
m1loss, m1accuracy = model.evaluate(X_test, y_test)
print("Test Loss:", m1loss)
print("Test Accuracy:", m1accuracy)
```

```python
from sklearn import metrics
y_pred = model.predict(X_test)
y_pred = np.round(y_pred).tolist()
print("accuracy score:{:.2f}%".format(metrics.accuracy_score(y_test, y_pred)*100))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
#print(cm)
import seaborn as sns
sns.heatmap(cm, annot=True, fmt='g')
```

```python
vgg16_X_train, vgg16_X_temp, vgg16_y_train, vgg16_y_temp = train_test_split(vgg16_X, y_label, test_size=0.3, random_state=55)
vgg16_X_val, vgg16_X_test, vgg16_y_val, vgg16_y_test = train_test_split(vgg16_X_temp, vgg16_y_temp, test_size=0.5, random_state=86)
print(vgg16_X_train.shape)
print(vgg16_y_train.shape)
print(vgg16_X_val.shape)
print(vgg16_y_val.shape)
print(vgg16_X_test.shape)
print(vgg16_y_test.shape)
```

```python
vgg16_weight_path = "C:/Users/ashwi/0 MV Assignment/Project/vgg16_weights_tf_dim_ordering_tf_kernels.h5"

base_model = tf.keras.applications.VGG16(
    weights="imagenet",
    include_top=False,
    input_shape=(224, 224, 3)
)

model1 = tf.keras.models.Sequential()
model1.add(base_model)
model1.add(tf.keras.layers.Flatten())
model1.add(tf.keras.layers.Dropout(0.5))
model1.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model1.layers[0].trainable = False

model1.compile(
    loss='binary_crossentropy',
    optimizer=tf.keras.optimizers.Adam(),
    metrics=['accuracy']
)

model1.summary()
```

```python
History1 = model1.fit(vgg16_X_train,vgg16_y_train, epochs = 5, validation_data = (vgg16_X_val,vgg16_y_val),batch_size=64, verbose = 1)
```

```python
m2loss, m2accuracy = model1.evaluate(vgg16_X_test, vgg16_y_test)
print("Test Loss:", m2loss)
print("Test Accuracy:", m2accuracy)
```

```python
from sklearn import metrics
y_pred = model1.predict(vgg16_X_test)
y_pred = np.round(y_pred).tolist()
print("accuracy score:{:.2f}%".format(metrics.accuracy_score(vgg16_y_test, y_pred)*100))
```

```python
base_model = tf.keras.applications.MobileNetV2(input_shape=(224,224,3),
                                               include_top=False,
                                               weights='imagenet')
```

```python
base_model.trainable = False
```

```python
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
prediction_layer = tf.keras.layers.Dense(1, activation='sigmoid')
```

```python
model3 = tf.keras.Sequential([
  base_model,
  global_average_layer,
  prediction_layer
])
```

```python
model3.compile(
    loss='binary_crossentropy',
    optimizer=tf.keras.optimizers.Adam(),
    metrics=['accuracy']
)

model3.summary()
```

```python
History3 = model3.fit(vgg16_X_train,vgg16_y_train, epochs = 10, validation_data = (vgg16_X_val,vgg16_y_val),batch_size=64, verbose = 1)
```
